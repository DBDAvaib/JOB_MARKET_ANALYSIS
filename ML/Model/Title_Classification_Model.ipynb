## One hot encoding for Location column

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
encoder.fit(df[["Location"]])  # Only fit on the intended column

# Transform and retrieve flat column names
transformed_values = encoder.transform(df[["Location"]])
column_names = encoder.get_feature_names_out(["Location"])

# Create DataFrame
df_transformed = pd.DataFrame(transformed_values, columns=column_names, index=df.index)


```

```python
df['Min Experience'] = df['Min Experience'].fillna(df['Min Experience'].mean())
df['Max Experience'] = df['Max Experience'].fillna(df['Max Experience'].mean())
```

```python
df['Location'] = df['Location'].fillna('Banglore')
```

```python
from sklearn.preprocessing import LabelEncoder

# create encoder
encoder = LabelEncoder()

# encode the values
df['Location'] = encoder.fit_transform(df['Location'])
```

```python
df['Min Salary'] = df['Min Salary'].fillna(df['Min Salary'].mean())
df['Max Salary'] = df['Max Salary'].fillna(df['Max Salary'].mean())
```

```python
df.drop(['Title','Location_nan'], axis=1 , inplace = True)
```

```python
x = df.drop(['Short_Title','Location_Banglore','Location_Chennai','Location_Delhi','Location_Gurugram','Location_Hyderabad','Location_Kolkata','Location_Mumbai','Location_Pune'],axis=1)

y = df['Short_Title']
```

```python
from sklearn.model_selection import train_test_split

# split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, train_size=0.8, random_state=123456
)
```

```python
from sklearn.linear_model import LogisticRegressionCV

def create_logistic_regression_model():

    # create model
    model = LogisticRegressionCV(max_iter=1000)

    # train the model
    model.fit(x_train, y_train)

    return model

from sklearn.neighbors import KNeighborsClassifier

def create_knn_model():

    # create the model
    model = KNeighborsClassifier(n_neighbors=5)

    # train the model
    model.fit(x_train, y_train)

    return model

from sklearn.svm import SVC

def create_svm_model():

    # create the model
    model = SVC(C=2, kernel='rbf')

    # train the model
    model.fit(x_train, y_train)

    return model

from sklearn.naive_bayes import GaussianNB

def create_naive_bayes_model():

    # create the model
    model = GaussianNB()

    # train the model
    model.fit(x_train, y_train)

    return model

from sklearn.tree import DecisionTreeClassifier

def create_decision_tree_model():

    # create the model
    model = DecisionTreeClassifier(max_depth=10)

    # train the model
    model.fit(x_train, y_train)

    return model

def create_xgboost():
    from xgboost import XGBClassifier
    # create the model
    model = XGBClassifier(n_estimators=35, max_depth=2, learning_rate=0.2)

    # train the model
    model.fit(x_train, y_train)

    return model

from catboost import CatBoostClassifier

def create_catboost_model():
    # create the model
    model = CatBoostClassifier(learning_rate=0.2)

    # train the model
    model.fit(x_train, y_train)

    return model


from sklearn.ensemble import AdaBoostClassifier

def create_adaboost_model():
    # create the model
    model = AdaBoostClassifier(n_estimators=35, learning_rate=0.2)

    # train the model
    model.fit(x_train, y_train)

    return model
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model):
    y_pred = model.predict(x_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred , average='micro')
    recall = recall_score(y_test, y_pred , average='micro')
    f1 = f1_score(y_test, y_pred , average='micro')
    return f"{accuracy:.2f}", f"{precision:.2f}", f"{recall:.2f}", f"{f1:.2f}"
```

```python
# create all the models
models = []
models.append({"name": "Logistic Regression", "model": create_logistic_regression_model()})
models.append({"name": "KNN", "model": create_knn_model()})
models.append({"name": "SVM", "model": create_svm_model()})
models.append({"name": "Naive Bayes", "model": create_naive_bayes_model()})
models.append({"name": "Decision Tree", "model": create_decision_tree_model()})
#models.append({"name": "Random Forest", "model": create_random_forest_model()})
models.append({"name": "XGBoost", "model":create_xgboost()})
models.append({"name": "AdaBoost", "model": create_adaboost_model()})
models.append({"name": "CatBoost", "model": create_catboost_model()})

# evaluate all the models
for info in models:
    accuracy, precision, recall, f1 = evaluate_model(info["model"])
    info["accuracy"] = accuracy
    info["precision"] = precision
    info["recall"] = recall
    info["f1"] = f1

# print the evaluation report
df_evaluation_report = pd.DataFrame(models, columns=["name", "accuracy","precision", "recall", "f1"])
df_evaluation_report
```
